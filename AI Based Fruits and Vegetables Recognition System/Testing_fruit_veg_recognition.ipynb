<h2 align="left"><font color=darkgreen>Test set Image Processing</font></h2>

```python
import tensorflow as tf  
test_set = tf.keras.utils.image_dataset_from_directory(
    'E:\\Fruits and Vegetables Recognition System\\Fruits and Vegetables Recognition\\test',
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(64, 64),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False
)
```

<div style="border-radius:10px; padding: 15px; font-size:100%; text-align:left">
<h2 align="left"><font color=green>Loading Model</font></h2>

```python
import numpy as np
import matplotlib as plt
cnn = tf.keras.models.load_model('E:\\Fruits and Vegetables Recognition System\\trained_model.h5')
```

<div style="border-radius:10px; padding: 15px; font-size:100%; text-align:left">
<h2 align="left"><font color=skyblue>Visualising and Performing Prediction on Single image
</font></h2>

```python
import cv2
import matplotlib.pyplot as plt

# Specify the full path to a specific image, not just the folder
image_path = 'E:\\Fruits and Vegetables Recognition System\\Fruits and Vegetables Recognition\\test\\peas\\Image_2.jpg'
# Reading the image
img = cv2.imread(image_path)

# Check if the image was successfully loaded
if img is None:
    print("Error: Image not found or the path is incorrect.")
else:
    # Converting BGR to RGB (since OpenCV uses BGR by default)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Displaying the image using matplotlib
    plt.imshow(img)
    plt.title('Test Image')
    plt.xticks([])  # Hides x-axis tick labels
    plt.yticks([])  # Hides y-axis tick labels
    plt.show()
```

<div style="border-radius:10px; padding: 15px; font-size:100%; text-align:left">
<h2 align="left"><font color=green>Testing Model</font></h2>



```python
image = tf.keras.preprocessing.image.load_img(image_path,target_size=(64,64))
input_arr = tf.keras.preprocessing.image.img_to_array(image)
input_arr = np.array([input_arr])  # Convert single image to a batch.
predictions = cnn.predict(input_arr)
```

```python
print(predictions)
```

```python
test_set.class_names
```

```python
result_index = np.argmax(predictions) #Return index of max element
print(result_index)
```

```python
# Displaying the image 
plt.imshow(img)
plt.title('Test Image')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
#Single image Prediction
print("It's a {}".format(test_set.class_names[result_index]))
```
